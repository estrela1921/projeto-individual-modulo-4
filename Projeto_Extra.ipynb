```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dados fornecidos
dados = {
    'ID da venda': [1, 2, 3, 4, 5],
    'Data da venda': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'ID do cliente': [101, 102, 103, 104, 105],
    'Valor da venda': [100.50, 150.75, 200.00, 120.25, 180.50],
    'Produto vendido': ['A', 'B', 'A', 'C', 'B'],
    'Categoria do produto': ['eletronicos', 'moda', 'eletronicos', 'casa', 'moda'],
    'Região de venda': ['Norte', 'Sul', 'Sudeste', 'Norte', 'Sul'],
    'Canal de venda': ['Online', 'Presencial', 'Presencial', 'Online', 'Online']
}

# Criar DataFrame
df = pd.DataFrame(dados)

# Exibir o DataFrame
print(df)
```

```python
df.describe(include='all')
```

##Qual a média dos valores das vendas?


```python
# Calcular a média dos valores das vendas
media_valores_vendas = df['Valor da venda'].mean()

# Exibir a média
print(f"A média dos valores das vendas é: {media_valores_vendas}")
```

```python
# Criar uma visualização de linha
plt.figure(figsize=(8, 4))
plt.plot(df['Data da venda'], df['Valor da venda'], marker='o', linestyle='-', color='b')
plt.title('Evolução dos Valores das Vendas ao Longo do Tempo')
plt.xlabel('Data da Venda')
plt.ylabel('Valor da Venda')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##Qual foi o total de vendas?

###O relatório fornece uma visão geral da quantidade de vendas realizadas pela empresa.
Permite identificar períodos de maior ou menor atividade de vendas.

```python
# Calcular o total de vendas
total_vendas = df['Valor da venda'].sum()

# Exibir o total de vendas
print(f"O total de vendas é: {total_vendas}")
```

```python
# Criar uma visualização de barras agrupadas
plt.figure(figsize=(10, 6))
sns.barplot(x='Categoria do produto', y='Valor da venda', hue='Região de venda', data=df, ci=None, palette='viridis')
plt.title('Total de Vendas por Categoria de Produto e Região de Venda')
plt.xlabel('Categoria do Produto')
plt.ylabel('Total de Vendas')
plt.legend(title='Região de Venda', loc='upper right')
plt.show()
```

##Qual o produto mais vendido?


###Analisar as categorias mais vendidas fornece insights sobre os produtos que têm maior aceitação no mercado.
Pode orientar decisões estratégicas, como investir em marketing para categorias específicas ou expandir a oferta de produtos populares.

```python
# Contar a ocorrência de cada produto
contagem_produtos = df['Produto vendido'].value_counts()

# Identificar o produto mais vendido (aquele com a contagem mais alta)
produto_mais_vendido = contagem_produtos.idxmax()

# Exibir o produto mais vendido
print(f"O produto mais vendido é: {produto_mais_vendido}")
```

```python
# Criar uma visualização de barras com cores diferentes e orientação horizontal
plt.figure(figsize=(10, 6))
sns.barplot(y=contagem_produtos.index, x=contagem_produtos.values, palette='muted', saturation=0.7)
plt.title('Contagem de Vendas por Produto')
plt.xlabel('Contagem de Vendas')
plt.ylabel('Produto Vendido')
plt.show()
```

##Qual a categoria com maior faturamento?


```python
# Agrupar por categoria e calcular a soma dos valores de venda
faturamento_por_categoria = df.groupby('Categoria do produto')['Valor da venda'].sum()

# Identificar a categoria com maior faturamento
categoria_maior_faturamento = faturamento_por_categoria.idxmax()

# Exibir a categoria com maior faturamento
print(f"A categoria com maior faturamento é: {categoria_maior_faturamento}")
```

```python
# Criar uma visualização de pizza
plt.figure(figsize=(6, 6))
plt.pie(faturamento_por_categoria, labels=faturamento_por_categoria.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set1.colors)
plt.title('Distribuição do Faturamento por Categoria de Produto')
plt.show()
```

##Qual o número de vendas online?


###Comparar as vendas entre canais presencial e online é crucial.
Identifica a preferência dos clientes em relação ao canal de compra.
Pode indicar a necessidade de ajustar estratégias de marketing, investir em presença online ou melhorar a experiência do cliente em lojas físicas

```python
# Contar a ocorrência de cada canal de venda
contagem_canal_venda = df['Canal de venda'].value_counts()

# Obter o número de vendas online
vendas_online = contagem_canal_venda.get('Online', 0)

# Exibir o número de vendas online
print(f"O número de vendas online é: {vendas_online}")
```

##Qual o número de vendas presenciais?


```python
# Contar a ocorrência de cada canal de venda
contagem_canal_venda = df['Canal de venda'].value_counts()

# Obter o número de vendas presenciais
vendas_presenciais = contagem_canal_venda.get('Presencial', 0)

# Exibir o número de vendas presenciais
print(f"O número de vendas presenciais é: {vendas_presenciais}")
```

```python
# Criar um DataFrame pivot para facilitar a visualização
pivot_df = df.pivot_table(index='Canal de venda', columns='Categoria do produto', values='Valor da venda', aggfunc='count', fill_value=0)

# Criar uma visualização de barras empilhadas
plt.figure(figsize=(10, 6))
pivot_df.plot(kind='bar', stacked=True, cmap='Set2')
plt.title('Contagem de Vendas por Canal de Venda e Categoria de Produto')
plt.xlabel('Canal de Venda')
plt.ylabel('Contagem de Vendas')
plt.legend(title='Categoria de Produto', loc='upper right')
plt.show()
```
